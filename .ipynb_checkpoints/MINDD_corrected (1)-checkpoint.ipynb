```python
def combine_agent_company(agent, company):
    if pd.isna(agent) and pd.isna(company):
        return "no_agent_company"
    elif pd.isna(agent):
        return company  # Retorna apenas o valor de company se agent estiver ausente
    elif pd.isna(company):
        return agent  # Retorna apenas o valor de agent se company estiver ausente
    else:
        return f"{agent}_{company}"  # Concatena se ambos estiverem presentes

# Aplicar a funÃ§Ã£o linha por linha
df['agent_company'] = df.apply(lambda row: combine_agent_company(row['agent'], row['company']), axis=1)

# Remover as colunas originais
df_cleaned = df.drop(columns=['agent', 'company'])
```

**Feature Selection**

1. Filter Methods

1.1 ANOVA

```python
continuous_columns = df_encoded.select_dtypes(include=['float64', 'int']).columns.tolist()
target_variable = 'is_canceled'
def FunctionAnova(inpData, TargetVariable, ContinuousPredictorList):
    SelectedPredictors = []
    
    print('##### ANOVA Results ##### \n')
    for predictor in ContinuousPredictorList:
        CategoryGroupLists = inpData.groupby(TargetVariable)[predictor].apply(list)
        AnovaResults = f_oneway(*CategoryGroupLists)
        
        if AnovaResults[1] < 0.05:
            print(f"{predictor} is correlated with {TargetVariable} | P-Value: {AnovaResults[1]}")
            SelectedPredictors.append(predictor)
        else:
            print(f"{predictor} is NOT correlated with {TargetVariable} | P-Value: {AnovaResults[1]}")
    
    return SelectedPredictors

selected_predictors_anova = FunctionAnova(df_encoded, target_variable, continuous_columns)
print("Selected Predictors (ANOVA):", selected_predictors_anova)
```

1.2 Chi-Square

```python
categorical_columns = df_encoded.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()

def FunctionChisq(inpData, TargetVariable, CategoricalVariablesList):
    SelectedPredictors = []

    for predictor in CategoricalVariablesList:
        CrossTabResult = pd.crosstab(index=inpData[TargetVariable], columns=inpData[predictor])
        ChiSqResult = chi2_contingency(CrossTabResult)
        
        if ChiSqResult[1] < 0.05:
            print(f"{predictor} is correlated with {TargetVariable} | P-Value: {ChiSqResult[1]}")
            SelectedPredictors.append(predictor)
        else:
            print(f"{predictor} is NOT correlated with {TargetVariable} | P-Value: {ChiSqResult[1]}")
    
    return SelectedPredictors

selected_predictors = FunctionChisq(df_encoded, target_variable, categorical_columns)
print("Selected Predictors:", selected_predictors)
```

2.1 LASSO Regularization

```python
def lasso_regularization(X_train, y_train):
    sel_ = SelectFromModel(LogisticRegression(C=0.5, penalty='l1', solver='liblinear', random_state=10))
    sel_.fit(X_train, y_train)

    selected_feat = X_train.columns[sel_.get_support()]
    removed_feats = X_train.columns[~sel_.get_support()]

    print("NÃºmero de caracterÃ­sticas cujo coeficiente foi reduzido a zero:", len(removed_feats))
    print("CaracterÃ­sticas removidas:", list(removed_feats))

    X_train_lasso = pd.DataFrame(sel_.transform(X_train), columns=selected_feat)
    
    return X_train_lasso, selected_feat

# Exemplo de uso com X_train e y_train jÃ¡ prÃ©-processados
X_lasso, selected_features = lasso_regularization(X_train, y_train)
print("Selected features:", selected_features)
```

```python
from keras.models import Sequential
from keras.layers import Dense
scaler = MinMaxScaler()

Xstd = scaler.fit_transform(X)

Xstd_train = scaler.fit_transform(X_train)
Xstd_test = scaler.fit_transform(X_test)
# Function to create MLP network 

def nnetClassif (inputDim, optimizer, neurons):
       
    # Initialize the Sequential model
    model = Sequential()
    
    model.add(Dense(units=neurons, 
                    input_dim=inputDim, 
                    kernel_initializer='uniform', 
                    activation='relu'))    
    
    model.add(Dense(units=neurons, 
                    kernel_initializer='uniform', 
                    activation='relu'))    
    
    model.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))  
    
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
            
    return model
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV


#for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

input_dim = Xstd_train.shape[1] 

classifModel = KerasClassifier(build_fn=nnetClassif, inputDim=input_dim, verbose=0)

# hyperparameter grid
Parameter_Trials={'model__optimizer': ['adam','rmsprop'],
                  'model__neurons': [8,16,32],
                  'batch_size': [30,45,60],
                      'epochs': [10,20]
                 }

# Apply GridSearchCV
grid_search = GridSearchCV(estimator=classifModel, param_grid=Parameter_Trials, scoring='accuracy', cv=3)
 
import time
StartTime=time.time()
 
# Running Grid Search for different parameters
grid_result = grid_search.fit(Xstd_train, y_train, validation_split=0.1, verbose=-1)

EndTime=time.time()
```

```python
print("Total Time: ", round((EndTime-StartTime)/60), 'min')
 
# Output the best parameters and accuracy
print(f"Best train accuracy: {round(grid_result.best_score_,3)} with best hyperparamters: \n {grid_result.best_params_}\n")

# After GridSearchCV, extract the best model
best_model = grid_result.best_estimator_.model_

# Print the summary of the best model
best_model.summary()
```

```python
best_model = grid_result.best_estimator_.model_
best_batch_size = grid_result.best_params_['batch_size']
best_epochs = grid_result.best_params_['epochs']

# Fit the best model on the training data
history = best_model.fit(Xstd_train, y_train, validation_split=0.1, 
                          batch_size=best_batch_size, 
                          epochs=best_epochs, 
                          verbose=0)
```

```python

# Naive Bayes
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Treinar o modelo Naive Bayes
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# PrevisÃµes e avaliaÃ§Ã£o
y_pred_nb = nb_model.predict(X_test)
y_prob_nb = nb_model.predict_proba(X_test)[:, 1]

print("Naive Bayes Classification Report:")
print(classification_report(y_test, y_pred_nb))

# Adicionar resultados para a curva ROC
fpr_nb, tpr_nb, _ = roc_curve(y_test, y_prob_nb)
plt.plot(fpr_nb, tpr_nb, label='Naive Bayes (AUC = %0.2f)' % roc_auc_score(y_test, y_prob_nb))
```

```python

# Support Vector Machines
from sklearn.svm import SVC

# Treinar o modelo SVM
svm_model = SVC(probability=True, kernel='rbf', random_state=42)
svm_model.fit(X_train, y_train)

# PrevisÃµes e avaliaÃ§Ã£o
y_pred_svm = svm_model.predict(X_test)
y_prob_svm = svm_model.predict_proba(X_test)[:, 1]

print("SVM Classification Report:")
print(classification_report(y_test, y_pred_svm))

# Adicionar resultados para a curva ROC
fpr_svm, tpr_svm, _ = roc_curve(y_test, y_prob_svm)
plt.plot(fpr_svm, tpr_svm, label='SVM (AUC = %0.2f)' % roc_auc_score(y_test, y_prob_svm))
```

```python

# CombinaÃ§Ã£o de modelos (Voting Classifier)
from sklearn.ensemble import VotingClassifier

# Criar o ensemble de modelos
ensemble_model = VotingClassifier(estimators=[
    ('rf', best_rf), 
    ('nb', nb_model),
    ('svm', svm_model)
], voting='soft')

ensemble_model.fit(X_train, y_train)

# PrevisÃµes e avaliaÃ§Ã£o
y_pred_ensemble = ensemble_model.predict(X_test)
y_prob_ensemble = ensemble_model.predict_proba(X_test)[:, 1]

print("Ensemble Model Classification Report:")
print(classification_report(y_test, y_pred_ensemble))

# Adicionar resultados para a curva ROC
fpr_ensemble, tpr_ensemble, _ = roc_curve(y_test, y_prob_ensemble)
plt.plot(fpr_ensemble, tpr_ensemble, label='Ensemble (AUC = %0.2f)' % roc_auc_score(y_test, y_prob_ensemble))
```
