<div class="alert alert-block alert-info">

Criou-se a coluna ['total_stays_nights'], que soma ['stays_in_weekend_nights'] e ['stays_in_week_nights'] para obter o total de noites de estadia.

Criou-se a coluna ['total_people'], que agrega ['adults'], ['children'] e ['babies'] para representar o total de pessoas na reserva.

As colunas que foram agrupadas foram removidas do DataFrame para evitar redundância e simplificar a análise.

##  3.2 Divisão dos dados, Escalonamento e PCA

O PCA (Principal Component Analysis) é uma técnica de redução de dimensionalidade que transforma variáveis correlacionadas num conjunto menor de variáveis não correlacionadas chamadas componentes principais. Estas componentes retêm a maior parte da variação dos dados, permitindo simplificar conjuntos de dados complexos, melhorar a eficiência dos modelos e lidar com multicolinearidade. Apesar de útil, o PCA pode reduzir a interpretabilidade das variáveis originais, já que as componentes principais são combinações lineares dessas variáveis.

## 3.3 Identificação e Remoção de outliers

Identificou-se os outliers através do Intervalo Interquartil.

```python
def calculate_outliers(df):
    outliers_info = {}
    for column in df.select_dtypes(include=['int64', 'float64']).columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        outliers_info[column] = len(outliers)
    
    return outliers_info

# Calcular os outliers em todas as variáveis numéricas
outliers_summary = calculate_outliers(df)
outliers_df = pd.DataFrame(list(outliers_summary.items()), columns=['Variable', 'Number of Outliers'])
outliers_df
```

# 4. Feature Selection

A seleção de características é uma etapa importante no processo de modelação, pois permite identificar e reter apenas as variáveis mais relevantes para a previsão, reduzindo a dimensionalidade dos dados e melhorando a eficiência dos modelos. Neste projeto, foram aplicados diferentes métodos de seleção de características, tanto de tipo filter como embedded, para garantir que as variáveis selecionadas contribuam de forma significativa para a performance do modelo.

## 4.1 Filter Methods

## 4.1.1 ANOVA

Utilizou-se o método ANOVA para medir a relação entre variáveis independentes contínuas e a variável target, ajudando a identificar quais variáveis têm um impacto estatisticamente significativo na previsão do cancelamento.

```python
# Define the ANOVA function
def FunctionAnova(inpData, TargetVariable, ContinuousPredictorList):
    from scipy.stats import f_oneway

    # Creating an empty list of final selected predictors
    SelectedPredictors = []
    
    print('##### ANOVA Results ##### \n')
    for predictor in ContinuousPredictorList:
        CategoryGroupLists = inpData.groupby(TargetVariable)[predictor].apply(list)
        AnovaResults = f_oneway(*CategoryGroupLists)
        
        # If the ANOVA P-Value is <0.05, that means we reject H0
        if (AnovaResults[1] < 0.05):
            print(predictor, 'is correlated with', TargetVariable, '| P-Value:', AnovaResults[1])
            SelectedPredictors.append(predictor)
        else:
            print(predictor, 'is NOT correlated with', TargetVariable, '| P-Value:', AnovaResults[1])
    
    return SelectedPredictors

# Create a list of all continuous predictors in df, excluding the target variable
ContinuousPredictorList = df.select_dtypes(include=['int64', 'float64']).columns.difference(['is_canceled']).tolist()

# Apply the function to the existing df
selected_predictors_all = FunctionAnova(df, 'is_canceled', ContinuousPredictorList)

# Print the selected predictors
print("Selected Predictors:", selected_predictors_all)
```

<div class="alert alert-block alert-info">
Através dos resultados da ANOVA, foi possível identificar as variáveis mais relevantes para a previsão de cancelamentos. Variáveis como adr, booking_changes, is_repeated_guest, lead_time, previous_bookings_not_canceled, previous_cancellations, required_car_parking_spaces, total_of_special_requests, total_people e total_stays_nights mostraram uma forte correlação com is_canceled, com valores-p extremamente baixos. Por outro lado, variáveis como arrival_date_week_number e days_in_waiting_list não apresentaram correlação significativa, evidenciada pelos seus valores-p elevados. 
</div>

## 4.1.2 CHI-SQUARED

Este teste foi aplicado para avaliar a dependência entre variáveis categóricas e a variável target, destacando quais atributos categóricos são mais importantes para a previsão.

```python
def FunctionChisq(inpData, TargetVariable, CategoricalVariablesList):
    from scipy.stats import chi2_contingency
    import pandas as pd
    
    # Creating an empty list of final selected predictors
    SelectedPredictors = []

    print('##### Chi-squared Test Results ##### \n')
    for predictor in CategoricalVariablesList:
        CrossTabResult = pd.crosstab(index=inpData[TargetVariable], columns=inpData[predictor])
        ChiSqResult = chi2_contingency(CrossTabResult)
        
        # If the ChiSq P-Value is <0.05, that means we reject H0
        if (ChiSqResult[1] < 0.05):
            print(predictor, 'is correlated with', TargetVariable, '| P-Value:', ChiSqResult[1])
            SelectedPredictors.append(predictor)
        else:
            print(predictor, 'is NOT correlated with', TargetVariable, '| P-Value:', ChiSqResult[1])        
            
    return SelectedPredictors

# Create a list of categorical predictors from the original DataFrame
CategoricalVariablesList = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Apply the function on the original DataFrame
selected_categorical_predictors = FunctionChisq(df, 'is_canceled', CategoricalVariablesList)

print("Selected Predictors:", selected_categorical_predictors)
```

<div class="alert alert-block alert-info"> 
Através dos resultados do teste chi-squared, foi possível identificar as variáveis categóricas que estão significativamente correlacionadas com is_canceled. Variáveis como hotel, meal, country, market_segment, distribution_channel, reserved_room_type, assigned_room_type, deposit_type, customer_type, reservation_status, reservation_status_date, agent_company e arrival_date_combined apresentaram valores-p extremamente baixos (próximos de zero), indicando uma forte correlação com a variável target. Estes resultados sugerem que estas características desempenham um papel importante na previsão de cancelamentos, justificando a sua inclusão na modelação.

## 4.2 Embedded Methods

## 4.2.1 LASSO Regularization

A LASSO (Least Absolute Shrinkage and Selection Operator) é uma técnica de regularização aplicada em modelos de regressão que adiciona uma penalidade à soma dos valores absolutos dos coeficientes. Essa penalização faz com que alguns coeficientes sejam reduzidos a zero, resultando em um modelo mais simples e parcimonioso, ideal para seleção de características.
A principal vantagem do LASSO é sua capacidade de realizar seleção de variáveis automaticamente, o que ajuda a reduzir o risco de sobreajustamento e a melhorar a interpretabilidade do modelo. No entanto, em situações em que existem muitas variáveis correlacionadas, o LASSO pode selecionar apenas uma delas e eliminar as demais, o que pode ser uma limitação em alguns casos

Utilizou-se este método, uma vez que, aplica penalizações que reduzem coeficientes de variáveis menos relevantes a zero, permitindo uma seleção automática de características. 

```python
def lasso_regularization(X_train_no_outliers, y_train_no_outliers, original_feature_names):
    sel_ = SelectFromModel(Lasso(alpha=0.1, random_state=10))
    sel_.fit(X_train_no_outliers, y_train_no_outliers)

    selected_indices = sel_.get_support(indices=True)  # Índices das características selecionadas
    selected_features = original_feature_names[selected_indices]  # Nomes das características selecionadas

    num_features_shrunk = np.sum(sel_.estimator_.coef_ == 0)
    print("Number of features whose coefficient was shrunk to zero:", num_features_shrunk)

    removed_feats = np.where(sel_.estimator_.coef_ == 0)[0]  # Índices das características removidas
    print("Features removed due to Lasso regularization (by index):", removed_feats)

    X_lasso = X_train_no_outliers[:, selected_indices]

    X_lasso_df = pd.DataFrame(X_lasso, columns=selected_features)

    print("\nTransformed DataFrame with selected features:")
    print(X_lasso_df.head())
    
    return X_lasso_df, selected_features

original_feature_names = X.columns 
X_lasso, selected_features = lasso_regularization(X_train_no_outliers, y_train_no_outliers, original_feature_names)

print("\nFinal DataFrame returned:")
print(X_lasso)

print("\nSelected features:")
print(selected_features)
```

<div class="alert alert-block alert-info">
Através da regularização LASSO, foi possível reduzir significativamente o número de características utilizadas no modelo. Dos 2.267 atributos originais, 2.252 tiveram os seus coeficientes reduzidos a zero, indicando que não contribuem de forma relevante para a previsão e foram removidos. Esta técnica ajudou a simplificar o modelo, deixando apenas as variáveis mais significativas.

O conjunto final inclui características como lead_time, previous_bookings_not_canceled, booking_changes, entre outras, que mostraram ter um impacto real na predição de is_canceled. O uso da regularização LASSO foi essencial para evitar o sobreajuste e melhorar a eficiência e a interpretabilidade do modelo, mantendo apenas as variáveis que mais contribuem para o desempenho preditivo.
</div>

## 5. Criação e Comparação entre modelos

**Decision Tree Classifier:** Este modelo de classificação organiza os dados numa estrutura de árvore, onde cada nó representa uma decisão baseada num atributo específico, e as folhas representam os resultados das classes. É fácil de interpretar e pode lidar com dados categóricos e numéricos, mas pode ser propenso a sobreajustamento, especialmente em árvores muito profundas.

**K-Nearest Neighbors (KNN):** É um algoritmo de classificação baseado em instâncias que atribui uma classe a um dado de teste com base nas classes dos seus vizinhos mais próximos. O parâmetro 'k' determina o número de vizinhos a considerar. KNN é simples e eficaz, mas pode ser computacionalmente caro com grandes volumes de dados e sensível a variáveis de escala.

**Support Vector Classifier (SVC):** Este modelo cria um hiperplano de separação no espaço de alta dimensionalidade para classificar os dados. O SVC é eficaz em casos de alta dimensionalidade e funciona bem com margens claras entre classes. Pode utilizar diferentes tipos de kernel para adaptar-se a distribuições complexas, mas o seu desempenho pode ser afetado por conjuntos de dados com sobreposição de classes.

**Naive Bayes:** É um classificador probabilístico baseado no teorema de Bayes, que assume independência condicional entre as características. Apesar da suposição de independência ser forte e raramente verdadeira, o Naive Bayes é eficiente e muitas vezes eficaz em problemas de classificação de texto e outros contextos com grande número de características. É particularmente rápido em treino e predição, mas pode ter desempenho limitado se a suposição de independência for muito violada.

Treinou-se os modelos Decision Tree Classifier, K-Nearest Neighbors, Support Vector Classifier e Naive Bayes.

```python
models = []
models.append(('CART', DecisionTreeClassifier()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('SVC', SVC()))
models.append(('Naive Bayes', GaussianNB()))
```

```python
metrics_columns = ['Model', 'Accuracy', 'Precision', 'Recall']
results_df = pd.DataFrame(columns=metrics_columns)
n_folds = 10

# Armazena os índices das características selecionadas no treinamento
selected_indices = np.where(np.isin(original_feature_names, selected_features))[0]

plt.figure(figsize=(10, 8))  # Cria uma nova figura para as curvas ROC

# Criar uma lista para armazenar os valores FPR, TPR e AUC de cada modelo
roc_data = []

for name, model in models:
    kfold = StratifiedKFold(n_splits=n_folds, shuffle=True, random_state=42)
    
    # Realizar a validação cruzada com os dados transformados
    cv_results = cross_val_score(model, X_lasso, y_train_no_outliers, cv=kfold, scoring='accuracy')
    
    # Treinar o modelo com os dados transformados
    model.fit(X_lasso, y_train_no_outliers)
    
    # Prever utilizando os dados de teste transformados
    X_test_lasso = X_test_scaled_pca[:, selected_indices]  # Ajuste para usar os índices selecionados
    
    # Previsões
    if hasattr(model, "predict_proba"):
        y_pred_proba = model.predict_proba(X_test_lasso)[:, 1]  # Probabilidade da classe positiva
    else:
        y_pred_proba = model.predict(X_test_lasso)  # Previsões diretas

    # Convertendo y_test para binário
    y_test_bin = (y_test > 1).astype(int)  # Ajuste esse critério conforme necessário
    
    # Calcular a matriz de confusão
    conf_matrix = confusion_matrix(y_test_bin, (y_pred_proba > 0.5).astype(int))
    conf_matrix_df = pd.DataFrame(conf_matrix, 
                                   index=['True Positives', 'True Negatives'], 
                                   columns=['Predicted Positives', 'Predicted Negatives'])

    # Exibir a matriz de confusão usando um gráfico
    plt.figure(figsize=(6, 4))
    sns.heatmap(conf_matrix_df, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {name}')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
    
    # Calcular a curva ROC
    fpr, tpr, _ = roc_curve(y_test_bin, y_pred_proba)
    roc_auc = auc(fpr, tpr)  # Calcula a AUC
    
    # Armazenar dados da ROC para plotagem posterior
    roc_data.append((fpr, tpr, roc_auc, name))

    # Calcular métricas
    accuracy = accuracy_score(y_test_bin, (y_pred_proba > 0.5).astype(int))
    precision = precision_score(y_test_bin, (y_pred_proba > 0.5).astype(int), average='weighted')
    recall = recall_score(y_test_bin, (y_pred_proba > 0.5).astype(int), average='weighted')

    # Usando pd.concat() em vez de append()
    results_df = pd.concat([results_df, pd.DataFrame([{
        'Model': name,
        'Accuracy': round(accuracy, 3),
        'Precision': round(precision, 3),
        'Recall': round(recall, 3)
    }])], ignore_index=True)

# Configurações do gráfico ROC
plt.figure(figsize=(10, 8))  # Nova figura para as curvas ROC
for fpr, tpr, roc_auc, name in roc_data:
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')  # Plota a curva ROC

plt.plot([0, 1], [0, 1], 'k--')  # Linha diagonal
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc='lower right')
plt.show()

print("\nMetrics Summary:")
print(results_df)
```

O modelo KNN (K-Nearest Neighbors) também demonstrou ser muito eficiente, com métricas de precisão, recall e AUC próximas de 0,99, o que mostra um bom equilíbrio entre a deteção de verdadeiros positivos e a minimização de falsos positivos e falsos negativos. ## 5.1 Neural Networks

```python
def nnetClassif(inputDim, optimizer='adam', neurons=32):
    model = Sequential()
    
    model.add(Dense(units=neurons, input_dim=inputDim, activation='relu', kernel_regularizer=l2(0.001)))
    model.add(Dropout(0.5))  

    model.add(Dense(units=neurons, activation='relu', kernel_regularizer=l2(0.001)))
    model.add(Dropout(0.5))  

    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
np.random.seed(42)
tf.random.set_seed(42)

input_dim = X_train_no_outliers.shape[1] 

classifModel = KerasClassifier(build_fn=nnetClassif, inputDim=input_dim, verbose=0)

Parameter_Trials={'model__optimizer': ['adam','rmsprop'],
                  'model__neurons': [8,16,32],
                  'batch_size': [30,45,60],
                      'epochs': [10,20]
                 }

grid_search = GridSearchCV(estimator=classifModel, param_grid=Parameter_Trials, scoring='accuracy', cv=3)
 
import time
StartTime=time.time()
 
grid_result = grid_search.fit(X_train_no_outliers, y_train_no_outliers, validation_split=0.1, verbose=-1)

EndTime=time.time()
```

```python
print("Total Time: ", round((EndTime-StartTime)/60), 'min')
 
# Output the best parameters and accuracy
print(f"Best train accuracy: {round(grid_result.best_score_,3)} with best hyperparamters: \n {grid_result.best_params_}\n")

# After GridSearchCV, extract the best model
best_model = grid_result.best_estimator_.model_

# Print the summary of the best model
best_model.summary()
```

## 6. Combinação de modelos

A combinação de modelos em aprendizagem automática é importante devido aos benefícios práticos e teóricos que proporciona na tarefa de previsão. Não existe um único algoritmo que tenha o melhor desempenho em todos os tipos de problemas; cada modelo explora diferentes representações e abordagens, resultando em desempenhos distintos. Desta forma, a combinação de modelos tira partido dessas variações para gerar resultados mais robustos e fiáveis.
 
Entre as principais vantagens da combinação de modelos destacam-se uma maior capacidade preditiva, uma vez que modelos combinados frequentemente superam o desempenho de modelos individuais ao explorar diferentes características dos dados e algoritmos. Além disso, oferece uma maior robustez, com previsões mais estáveis, já que os erros dos modelos individuais podem ser compensados por outros modelos na combinação. Outro ponto importante é a redução do viés e da variância, uma vez que técnicas como bagging e boosting ajudam a diminuir a variância e o viés, melhorando a capacidade de generalização dos modelos.
 
Assim, a estratégia de combinação de modelos é uma abordagem eficaz para maximizar o desempenho em tarefas de classificação e regressão, explorando a diversidade de métodos para criar sistemas preditivos mais eficientes e resilientes. 

```python
knn_model = next((model for name, model in models if name == 'KNN'), None)
svc_model = next((model for name, model in models if name == 'SVC'), None)

# Criando o Voting Classifier com KNN e SVC
voting_clf = VotingClassifier(
    estimators=[
        ('KNN', knn_model), 
        ('SVC', svc_model)  # SVC precisa ter probability=True para votação suave
    ],
    voting='soft'  # Usando votação suave
)

# Treinando o Voting Classifier
voting_clf.fit(X_train_no_outliers, y_train_no_outliers)

# Fazendo previsões
y_pred_voting = voting_clf.predict(X_test_no_outliers)

# Avaliando o modelo
print("Voting Classifier Performance:")
print(classification_report(y_test_no_outliers, y_pred_voting))
```

```python
final_model = LogisticRegression()

# Criando o Stacking Classifier
stacking_clf = StackingClassifier(
    estimators=models,
    final_estimator=final_model
)

# Treinando o Stacking Classifier
stacking_clf.fit(X_train_no_outliers, y_train_no_outliers)

# Fazendo previsões
y_pred_stacking = stacking_clf.predict(X_test_no_outliers)

# Avaliando o modelo
print("Stacking Classifier Performance:")
print(classification_report(y_test_no_outliers, y_pred_stacking))
```
